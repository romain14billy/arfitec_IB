```python
# =====================================================
# CHOIX DE L'EXPÉRIENCE
# =====================================================

EXPERIMENT = "exp_global_chopper"
```

```python
import pickle
import sys
from pathlib import Path

# =====================================================
# CONFIGURATION DU PROJET
# =====================================================

PROJECT_ROOT = Path.cwd().resolve().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from src.analysis.fonction import process_experiment
from src.utils.config_loader import load_config

# =====================================================
# CHARGEMENT DE LA CONFIGURATION ET DES EXPÉRIENCES
# =====================================================

CONFIG = load_config(EXPERIMENT)

experiments_path = PROJECT_ROOT / "experiments" / EXPERIMENT / "experiments.pkl"
with experiments_path.open("rb") as f:
    experiments = pickle.load(f)

# =====================================================
# CALCUL DES RÉSULTATS
# =====================================================

results = {
    exp.name: process_experiment(exp, CONFIG)
    for exp in experiments
}

results_path = PROJECT_ROOT / "experiments" / EXPERIMENT / "results.pkl"
with results_path.open("wb") as f:
    pickle.dump(results, f)

print(f"{len(results)} résultats sauvegardés dans {results_path}")
```
