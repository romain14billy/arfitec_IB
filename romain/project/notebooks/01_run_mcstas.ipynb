```python
# =====================================================
# CHOIX DE L'EXPÉRIENCE
# =====================================================

EXPERIMENT = "exp_detect_peak"
```

```python
import pickle
import sys
from pathlib import Path

# =====================================================
# CONFIGURATION DU PROJET
# =====================================================

PROJECT_ROOT = Path.cwd().resolve().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from src.mcstas.mcstas_runner import (
    build_experiments,
    clean_project,
    compile_mcstas,
    run_campaign,
)
from src.utils.config_loader import load_config

# =====================================================
# CHARGEMENT DE LA CONFIGURATION
# =====================================================

config = load_config(EXPERIMENT)
print(f"Experiment: {config['name']}")

# =====================================================
# EXÉCUTION DE LA CAMPAGNE MCSTAS
# =====================================================

compile_mcstas(config)
run_campaign(config)

experiments = build_experiments(config)
print(f"\n{len(experiments)} expériences générées")

# =====================================================
# SAUVEGARDE DES RÉSULTATS
# =====================================================

output_file = PROJECT_ROOT / "experiments" / EXPERIMENT / "experiments.pkl"
output_file.parent.mkdir(parents=True, exist_ok=True)

with output_file.open("wb") as f:
    pickle.dump(experiments, f)

# =====================================================
# NETTOYAGE DU DOSSIER NOTEBOOKS
# =====================================================

clean_project(PROJECT_ROOT / "notebooks")
```
